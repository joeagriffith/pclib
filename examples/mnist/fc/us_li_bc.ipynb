```python
import torch
import torch.nn.functional as F
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline

from pclib.nn.models import FCClassifierUsLiBc
from pclib.optim.train import train
from pclib.optim.eval import track_vfe, accuracy
from pclib.utils.functional import format_y, reTanh, identity
from pclib.utils.customdataset import PreloadedDataset
```

```python
seed = 42
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
torch.manual_seed(seed)
class TanhTransform(object):
    def __init__(self, a=1., b=0., c=1.0):
        self.a = a
        self.b = b
        self.c = c

    def __call__(self, img):
        return ((img * self.a).tanh() + self.b) * self.c

class InvTanhTransform(object):
    def __call__(self, img):
        num = 1 + img
        div = (1 - img).clamp(min=1e-6)
        m = 0.5 * torch.log(num / div)
        return m

class SigmoidTransform(object):
    def __call__(self, img):
        return img.sigmoid()
    
class ReLUTanhTransform(object):
    def __call__(self, img):
        return F.relu(img.tanh())

class ThresholdTransform(object):
    def __init__(self, threshold=0.5):
        self.threshold = threshold

    def __call__(self, img):
        return (img > self.threshold).float()

train_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.RandomAffine(degrees=10, translate=(0.1, 0.1), scale=(0.9, 1.1)),                                
    # ThresholdTransform(0.8),
    transforms.Normalize((0.1307,), (0.3081,)),
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    # ThresholdTransform(0.9),
    # InvTanhTransform(),
    transforms.Normalize((0.1307,), (0.3081,)),
    # SigmoidTransform(),
    # TanhTransform(a=2.0),
    # TanhTransform(a=1.0, b=1.0, c=0.5),
])

dataset = datasets.MNIST('../Datasets/', train=True, download=False, transform=transforms.ToTensor())
# shorten dataset
# length = 10
# dataset = torch.utils.data.Subset(dataset, range(length))
# train_dataset = PreloadedDataset.from_dataset(dataset, train_transform, device)

VAL_RATIO = 0.2
val_len = int(len(dataset) * VAL_RATIO)
train_len = len(dataset) - val_len
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_len, val_len])
train_dataset = PreloadedDataset.from_dataset(train_dataset, train_transform, device)
val_dataset = PreloadedDataset.from_dataset(val_dataset, val_transform, device)
INPUT_SHAPE = 784
NUM_CLASSES = 10
```

```python
# Show some images
fig, axes = plt.subplots(1, 10, figsize=(15, 5))
for i, ax in enumerate(axes):
    img, label = train_dataset[i]
    ax.imshow(img.squeeze().cpu(), cmap='gray')
    ax.axis('off')
plt.show()

# print max min values
print('Max value: ', train_dataset.transformed_images.max())
print('Min value: ', train_dataset.transformed_images.min())
```

```python
torch.manual_seed(seed)

model_name = 'Us-Li-Bc'
model = FCClassifierUsLiBc(
    in_features=INPUT_SHAPE, 
    num_classes=NUM_CLASSES, 
    hidden_sizes=[200, 200],
    bias=True, 
    symmetric=True,
    precision_weighted=False,
    actv_fn=identity,
    steps=60,
    gamma=0.1,
    ).to(device)
```

```python
# Train Loop
NUM_EPOCHS = 30
BATCH_SIZE = 1000

log_dir = f'examples/mnist/logs/architecture/{model_name}'
train(
    model, 
    train_dataset, 
    val_dataset, 
    NUM_EPOCHS, 
    lr=0.001, 
    c_lr=0.0,
    batch_size=BATCH_SIZE,
    reg_coeff=1.0,
    optim='AdamW',
    save_best=False,
    log_dir=log_dir,
)
NUM_EPOCHS = 10
train(
    model, 
    train_dataset, 
    val_dataset, 
    NUM_EPOCHS, 
    lr=0.0,
    c_lr=0.01,
    batch_size=BATCH_SIZE,
    reg_coeff=0.02,
    optim='AdamW',
    save_best=False,
    log_dir=log_dir,
)
NUM_EPOCHS = 10
train(
    model, 
    train_dataset, 
    val_dataset, 
    NUM_EPOCHS, 
    lr=0.0,
    c_lr=0.001,
    batch_size=BATCH_SIZE,
    reg_coeff=0.02,
    optim='AdamW',
    save_best=False,
    log_dir=log_dir,
)
```
