```python
import torch
import torch.nn.functional as F
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline

from pclib.nn.models import FCClassifier
from pclib.nn.layers import FCPW
from pclib.optim.train import train
from pclib.optim.eval import track_vfe, accuracy
from pclib.utils.functional import format_y, reTanh
from pclib.utils.customdataset import PreloadedDataset
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python

class TanhTransform(object):
    def __init__(self, a=1., b=0., c=1.0):
        self.a = a
        self.b = b
        self.c = c

    def __call__(self, img):
        return ((img * self.a).tanh() + self.b) * self.c

class InvTanhTransform(object):
    def __call__(self, img):
        num = 1 + img
        div = (1 - img).clamp(min=1e-6)
        m = 0.5 * torch.log(num / div)
        return m

class SigmoidTransform(object):
    def __call__(self, img):
        return img.sigmoid()
    
class ReLUTanhTransform(object):
    def __call__(self, img):
        return F.relu(img.tanh())

train_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    transforms.RandomAffine(degrees=10, translate=(0.1, 0.1), scale=(0.9, 1.1)),                                
    # transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    # TanhTransform(a=0.5),
    TanhTransform(a=1.0, b=1.0, c=0.5),
])

val_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    # InvTanhTransform(),
    # SigmoidTransform(),
    # TanhTransform(a=0.5),
    TanhTransform(a=1.0, b=1.0, c=0.5),
])

dataset = datasets.CIFAR10('../Datasets/', train=True, download=True, transform=transforms.ToTensor())
# # shorten dataset
# length = 1000
# dataset = torch.utils.data.Subset(dataset, range(length))

VAL_RATIO = 0.2
val_len = int(len(dataset) * VAL_RATIO)
train_len = len(dataset) - val_len
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_len, val_len])
train_dataset = PreloadedDataset.from_dataset(train_dataset, train_transform, device)
val_dataset = PreloadedDataset.from_dataset(val_dataset, val_transform, device)
INPUT_SHAPE = 3072
NUM_CLASSES = 10
```

```python
# Show some images
fig, axes = plt.subplots(1, 10, figsize=(15, 5))
for i, ax in enumerate(axes):
    img, label = train_dataset[i]
    ax.imshow(img.squeeze().cpu().permute(1, 2, 0))
    ax.axis('off')
plt.show()

# print max min values
print('Max value: ', train_dataset.transformed_images.max())
print('Min value: ', train_dataset.transformed_images.min())
print('Data shape: ', train_dataset.transformed_images.shape)
```

```python
seed = 42
torch.manual_seed(seed)

model_name = 'fc_classifierLowlr'
model = FCClassifier(
    in_features=INPUT_SHAPE, 
    num_classes=NUM_CLASSES, 
    hidden_sizes=[2000, 2000, 2000],
    bias=True, 
    symmetric=True, 
    precision_weighted=False,
    actv_fn=F.tanh,
    steps=100,
    gamma=0.1,
    ).to(device)
```

```python
# Train Loop
NUM_EPOCHS = 250
BATCH_SIZE = 1000
REG_COEFF = 1.0

train(
    model, 
    model_name,
    train_dataset, 
    val_dataset, 
    NUM_EPOCHS, 
    lr=0.00001,
    c_lr=0.0,
    batch_size=BATCH_SIZE,
    reg_coeff=REG_COEFF,
    optim='AdamW',
    grad_mode='manual',
    save_best=False,
)
LEARNING_RATE = 0.0001
NUM_EPOCHS = 30
train(
    model, 
    model_name,
    train_dataset, 
    val_dataset, 
    NUM_EPOCHS, 
    lr=0.000001,
    c_lr=0.01,
    batch_size=BATCH_SIZE,
    reg_coeff=REG_COEFF,
    stats=stats,
    optim='AdamW',
)
```
